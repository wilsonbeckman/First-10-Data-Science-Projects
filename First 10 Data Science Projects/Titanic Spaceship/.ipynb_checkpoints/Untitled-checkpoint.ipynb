```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid', {"grid.color": ".6", "grid.linestyle": ":"})
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
#PassengerId - A unique Id for each passenger.
#HomePlanet - The planet the passenger departed from, typically their planet of permanent residence.
#CryoSleep - Indicates whether the passenger elected to be put into suspended animation for the duration of the voyage.
#Cabin - The cabin number where the passenger is staying.
#Destination - The planet the passenger will be debarking to.
#Age - The age of the passenger.
#VIP - Whether the passenger has paid for special VIP service during the voyage.
#RoomService - Amount the passenger has billed for room service.
#FoodCourt - Amount the passenger has billed at the food court.
#Shopping Mall - Amount the passenger has billed at the shopping mall.
#Spa - Amount the passenger has billed at the spa.
#VRDeck - Amount the passenger has billed at the VR deck.
#Name - The name of the passenger.
#Transported - Whether the passenger was transported to another dimension.
```

```python
train['ID1'] = train['PassengerId'].str[:4]
test['ID1'] = test['PassengerId'].str[:4]
```

```python
train['ID2'] = train['PassengerId'].str[-2:]
test['ID2'] = test['PassengerId'].str[-2:]
```

```python
train['ID2'] = train['ID2'].astype('int')
test['ID2'] = test['ID2'].astype('int')
```

```python
train['With_family'] = [1 if x >= 2 else 0 for x in train['ID2']]
test['With_family'] = [1 if x >= 2 else 0 for x in test['ID2']]
```

```python
spend_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for x in spend_cols:
    train[x] = train[x].fillna(0)
    test[x] = test[x].fillna(0)
```

```python
train['TotalSpend'] = train['RoomService'] + train['FoodCourt'] + train['ShoppingMall'] + train['Spa'] + train['VRDeck']
test['TotalSpend'] = test['RoomService'] + test['FoodCourt'] + test['ShoppingMall'] + test['Spa'] + test['VRDeck']
```

```python
train = pd.get_dummies(data = train, columns = ['HomePlanet'])
test = pd.get_dummies(data = test, columns = ['HomePlanet'])
```

```python
train = pd.get_dummies(data = train, columns = ['Destination'])
test = pd.get_dummies(data = test, columns = ['Destination'])
```

```python
train['Cabin'].unique() #Use movies['rating'] = pd.Categorical(movies['rating'], ['NR', 'G', 'PG', 'PG-13', 'R'], ordered = True)
```

```python
#split_df = df['exerciseDescription'].str.split('(\d+)', expand=True)
#df.exercise = split_df[0].replace('[\- ]', '', regex=True)
cabin_split = train['Cabin'].str.split('(\d+)', expand=True)
train['Cabin_first_letter'] = cabin_split[0].replace('[/\- ]', '', regex = True)
train['Cabin_second_letter'] = cabin_split[2].replace('[/\- ]', '', regex = True)
train['Cabin_Number'] = cabin_split[1]

cabin_split2 = test['Cabin'].str.split('(\d+)', expand=True)
test['Cabin_first_letter'] = cabin_split[0].replace('[/\- ]', '', regex = True)
test['Cabin_second_letter'] = cabin_split[2].replace('[/\- ]', '', regex = True)
test['Cabin_Number'] = cabin_split[1]
```

```python
alphabet = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'H': 8, 'I': 9, 'J': 10, 'K': 11, 'L': 12, 'M': 13, 'N': 14, 'O': 15, 'P': 16, 'Q': 17, 'R': 18, 'S': 19, 'T': 20, 'U': 21, 'V': 22, 'W': 23, 'X': 24, 'Y': 25, 'Z': 26}
train['Cabin_first_letter_ordered'] = train['Cabin_first_letter'].replace(alphabet)
train['Cabin_second_letter_ordered'] = train['Cabin_second_letter'].replace(alphabet)
test['Cabin_first_letter_ordered'] = test['Cabin_first_letter'].replace(alphabet)
test['Cabin_second_letter_ordered'] = test['Cabin_second_letter'].replace(alphabet)
```

```python
name_split = train['Name'].str.split(' ')
train['last_name'] = name_split.str.get(-1)
name_split2 = test['Name'].str.split(' ')
test['last_name'] = name_split2.str.get(-1)
```

```python
a = train['last_name']
b = test['last_name']

import collections
duplicated_train = [item for item, count in collections.Counter(a).items() if count > 1]
duplicated_test = [item for item, count in collections.Counter(b).items() if count > 1]
```

```python
train['With_family2'] = [1 if x in duplicated_train else 0 for x in train['last_name']]
test['With_family2'] = [1 if x in duplicated_test else 0 for x in test['last_name']]
```

```python
bool_cols = ['CryoSleep', 'VIP', 'Transported']
for i in bool_cols:
    train[i] = train[i].replace({True: 1, False: 0})
test_bool_cols = ['CryoSleep', 'VIP']
for i in test_bool_cols:
    test[i] = test[i].replace({True: 1, False: 0})
```

```python
cols = ['CryoSleep', 'VIP']
for col in cols:
    train[col] = train[col].fillna(0)
    test[col] = test[col].fillna(0)
```

```python
train['Age'] = train['Age'].fillna(train['Age'].mean())
test['Age'] = test['Age'].fillna(test['Age'].mean())
```

```python
train = train.dropna()
test = test.dropna()
```

```python
train['ID1'] = train['ID1'].astype('int')
test['ID1'] = train['ID1'].astype('int')
train['Cabin_Number'] = train['Cabin_Number'].astype('int')
test['Cabin_Number'] = train['Cabin_Number'].astype('int')
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier

from sklearn.model_selection import cross_val_score
```

```python
trainML = train[['RoomService', 'Spa', 'VRDeck', 'TotalSpend', 'HomePlanet_Earth', 'Cabin_first_letter_ordered', 'Cabin_second_letter_ordered', 'Destination_55 Cancri e', 'HomePlanet_Europa', 'CryoSleep', 'Transported']]
testML = test[['PassengerId', 'RoomService', 'Spa', 'VRDeck', 'TotalSpend', 'HomePlanet_Earth', 'Cabin_first_letter_ordered', 'Cabin_second_letter_ordered', 'Destination_55 Cancri e', 'HomePlanet_Europa', 'CryoSleep']]
trainML.head()
```

```python
X = trainML.drop('Transported', axis = 1)
y = trainML['Transported']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 30)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
rf_train_score = model.score(X_train, y_train)
rf_test_score = model.score(X_test, y_test)
print('Train Score: ', rf_train_score)
print('Test Score: ', rf_test_score)
```

```python
model = GradientBoostingClassifier()
model.fit(X_train, y_train)
gb_train_score = model.score(X_train, y_train)
gb_test_score = model.score(X_test, y_test)
print('Train Score: ', gb_train_score)
print('Test Score: ', gb_test_score)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
lr_train_score = model.score(X_train, y_train)
lr_test_score = model.score(X_test, y_test)
print('Train Score: ', lr_train_score)
print('Test Score: ', lr_test_score)
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
knn_train_score = model.score(X_train, y_train)
knn_test_score = model.score(X_test, y_test)
print('Train Score: ', knn_train_score)
print('Test Score: ', knn_test_score)
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
gnb_train_score = model.score(X_train, y_train)
gnb_test_score = model.score(X_test, y_test)
print('Train Score: ', gnb_train_score)
print('Test Score: ', gnb_test_score)
```

```python
model = Perceptron()
model.fit(X_train, y_train)
per_train_score = model.score(X_train, y_train)
per_test_score = model.score(X_test, y_test)
print('Train Score: ', per_train_score)
print('Test Score: ', per_test_score)
```

```python
model = SGDClassifier()
model.fit(X_train, y_train)
sgd_train_score = model.score(X_train, y_train)
sgd_test_score = model.score(X_test, y_test)
print('Train Score: ', sgd_train_score)
print('Test Score: ', sgd_test_score)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
dt_train_score = model.score(X_train, y_train)
dt_test_score = model.score(X_test, y_test)
print('Train Score: ', dt_train_score)
print('Test Score: ', dt_test_score)
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
xgb_train_score = model.score(X_train, y_train)
xgb_test_score = model.score(X_test, y_test)
print('Train Score: ', xgb_train_score)
print('Test Score: ', xgb_test_score)
```

```python
model = RandomForestClassifier(random_state = 42)
param_grid = {'n_estimators': [50, 100, 200, 500, 1000], 'max_features': ['auto', 'sqrt', 'log2'], 'max_depth' : [3,4,5,6,7,8,9,10], 'criterion' :['gini', 'entropy']}
CV_model = GridSearchCV(estimator = model, param_grid = param_grid, cv = 5)
CV_model.fit(X_train, y_train)
```
