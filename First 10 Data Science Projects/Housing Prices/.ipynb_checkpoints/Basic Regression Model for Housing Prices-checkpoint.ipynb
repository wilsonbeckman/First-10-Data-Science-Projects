```python
#I assumed if the column was filled in the correct data point is 0 or NA
train['LotFrontage'].fillna(0, inplace = True)
train['Alley'].fillna('NA', inplace = True)
train['MasVnrType'].fillna('None', inplace = True)
train['MasVnrArea'].fillna(0, inplace = True)
train['BsmtQual'].fillna('NA', inplace = True)
train['BsmtCond'].fillna('NA', inplace = True)
train['BsmtExposure'].fillna('NA', inplace = True)
train['BsmtFinType1'].fillna('NA', inplace = True)
train['BsmtFinType2'].fillna('NA', inplace = True)
train['Electrical'].fillna('Mix', inplace = True)
train['FireplaceQu'].fillna('NA', inplace = True)
train['GarageType'].fillna('NA', inplace = True)
train['GarageYrBlt'].fillna(train['GarageYrBlt'].min(), inplace = True)
train['GarageFinish'].fillna('NA', inplace = True)
train['GarageQual'].fillna('NA', inplace = True)
train['GarageCond'].fillna('NA', inplace = True)
train['PoolQC'].fillna('NA', inplace = True)
train['Fence'].fillna('NA', inplace = True)
train['MiscFeature'].fillna('NA', inplace = True)
```

```python
#Created a couple features that could potentially result in information gain
train['Attic'] = [1 if x == 40 else 0 for x in train['MSSubClass']]
train['Unfinished'] = [1 if x == 45 else 0 for x in train['MSSubClass']]
train['Split'] = [1 if x == 80 or x== 85 or x == 180 else 0 for x in train['MSSubClass']]
train['PUD'] = [1 if x == 120 or x== 150 or x == 160 else 0 for x in train['MSSubClass']]
```

```python
#One Hot Encode Categorical Data
train = pd.get_dummies(data = train, columns = ['MSZoning', 'Street', 'LandContour', 'LotConfig', 'Neighborhood', 
                                                'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'OverallQual', 
                                                'OverallCond', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 
                                               'MasVnrType', 'Foundation', 'Heating', 'Electrical', 'Functional', 
                                               'GarageType', 'PavedDrive', 'Fence', 'MiscFeature', 'SaleType', 
                                                'SaleCondition'])
```

```python
#Label Encoded Data
#I found the .replace function to be the easiest way to Label Encode Orderly data since there were so many categorical features
train['Alley'] = [0 if x == 'NA' else 0 for x in train['Alley']]
train['LotRegular'] = [1 if x == 'Regular' or x == 'Slightly irregular' else 0 for x in train['LotShape']]
train['Utilities'] = train['Utilities'].replace({'AllPub': 3, 'NoSewr': 2, 'NoSeWa': 1, 'ELO': 0})
train['LandSlope'] = train['LandSlope'].replace({'Gtl': 0, 'Mod': 1, 'Sev': 2})
train['ExterQual'] = train['ExterQual'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
train['ExterCond'] = train['ExterCond'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
train['BsmtQual'] = train['BsmtQual'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0})
train['BsmtCond'] = train['BsmtCond'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0})
train['BsmtExposure'] = train['BsmtExposure'].replace({'Gd': 4, 'Av': 3, 'Mn': 2, 'No': 2, 'NA': 1, })
train['BsmtFinType1'] = train['BsmtFinType1'].replace({'GLQ': 6, 'ALQ': 5, 'BLQ': 4, 'Rec': 3, 'LwQ': 2, 'Unf': 1, 'NA': 0})
train['BsmtFinType2'] = train['BsmtFinType2'].replace({'GLQ': 6, 'ALQ': 5, 'BLQ': 4, 'Rec': 3, 'LwQ': 2, 'Unf': 1, 'NA': 0})
train['HeatingQC'] = train['HeatingQC'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
train['CentralAir'] = train['CentralAir'].replace({'Yes': 1, 'No': 0})
train['KitchenQual'] = train['KitchenQual'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
train['FireplaceQu'] = train['FireplaceQu'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0})
train['GarageFinish'] = train['GarageFinish'].replace({'Fin': 3, 'RFn': 2, 'Unf': 1, 'NA': 0})
train['GarageQual'] = train['GarageQual'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0})
train['GarageCond'] = train['GarageCond'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0})
train['PoolQC'] = train['PoolQC'].replace({'Ex': 4, 'Gd': 3, 'TA': 2, 'Fa': 1, 'NA': 0})
train['CentralAir'] = train['CentralAir'].replace({'Y':1, 'N': 0})
```

```python
train = train.drop(['Id', 'MSSubClass', 'LotShape', 'LotRegular', 'Alley'], axis = 1)
```

```python
#Normalized Data
normalizer = tf.keras.layers.Normalization(axis = -1)
```

```python
normalizer.adapt(X)
```

```python
#Mean of features when normalized
print(normalizer.mean.numpy())
```

```python
#comparison between regular data and normalized data
first = np.array(X[:1])

with np.printoptions(precision=2, suppress=True):
    print('First example:', first)
    print()
    print('Normalized:', normalizer(first).numpy())
```

```python
def build_and_compile_model(normalizer):
    model = tf.keras.Sequential([
        normalizer,
        keras.layers.Dense(2048, activation = 'relu'), 
        keras.layers.Dense(1024, activation = 'relu'),
        keras.layers.Dense(512, activation = 'relu'), 
        keras.layers.Dense(256, activation = 'relu'),
        keras.layers.Dense(128, activation = 'relu'), 
        keras.layers.Dense(64, activation = 'relu'), 
        keras.layers.Dense(1, activation = 'relu')
    ])
    model.compile(loss = 'mean_absolute_error', 
                 optimizer = tf.keras.optimizers.Adam(0.01))
    return model
```

```python
first_model = build_and_compile_model(normalizer)
```

```python
first_model.summary()
```

```python
#Used 50 epochs here since epochs over 50 tend to drasically overfit the model
%%time
history = first_model.fit(
    X,
    y,
    validation_split=0.2,
    verbose=0, 
    epochs=50)
```

```python
#plotted training and validation loss to show the difference over time
plt.plot(np.arange(0, 50), history.history['loss'], label = 'loss')
plt.plot(np.arange(0, 50), history.history['val_loss'], label = 'validation loss')
plt.ylim([0, 40000])
plt.axhline(y=np.min(history.history['loss']), xmin=0, xmax=1)
plt.axhline(y=np.min(history.history['val_loss']), xmin=0, xmax=1)
plt.xlabel('Epoch')
plt.ylabel('Mean Absolute Error')
plt.legend()
plt.show()
```

```python
print('Lowest Mean Absolute Error for the training set', np.min(history.history['loss']))

print('Lowest Mean Absolute Error for the validation set', np.min(history.history['val_loss']))
```

I tried several different model size here from a model that varied in depth and width. The best scenario in terms of validation loss was a 6 layer model with an extremely large funnel. 

I orignally ran a model that included all features and saw slight improvement when I reduced the model by the 100 least influencial features in terms of the feature's correlation coefficient to the Sale Price. 

In almost every scenario the model overfits. The obvious issue here is there isn't enough data, especially since there's a lot of features. We'll need more data to improve the model bias. 

The model featured above proved to be the best in terms of both validation and training loss. 

The model was on average 8800 dollars off the actual sales price on the training set and 18900 dollars off the actual sales price on the validation set
