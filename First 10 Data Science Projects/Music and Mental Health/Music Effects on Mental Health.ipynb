```python
cols = ['Frequency [Classical]', 'Frequency [Country]', 'Frequency [EDM]', 'Frequency [Folk]', 'Frequency [Gospel]', 'Frequency [Hip hop]', 'Frequency [Jazz]', 'Frequency [K pop]', 'Frequency [Latin]', 'Frequency [Lofi]', 'Frequency [Metal]', 'Frequency [Pop]', 'Frequency [R&B]', 'Frequency [Rap]', 'Frequency [Rock]', 'Frequency [Video game music]']
for col in cols:
    df[col] = df[col].replace({'Very frequently': 3, 'Sometimes': 2, 'Rarely': 1, 'Never': 0})
```

```python
df['Music effects'] = df['Music effects'].replace({'Improve': 1, 'No effect': 0, 'Worsen': -1})
```

```python
cols = ['While working', 'Instrumentalist', 'Composer', 'Exploratory', 'Foreign languages']
for col in cols:
    df[col] = df[col].replace({'Yes': 1, 'No': 0})
df.head()
```

I converted some of the categorical data to be numerical to build out a correlation matrix and dive deeper into the data

```python
corr_matrix = df.corr()
plt.figure(figsize = (16, 16))
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

```python
corr_matrix['Music effects'].sort_values(ascending = False)[1:]
```

As expected, there appears to be a slight negative correlation between Age and Music effects. 

Surprisingly, respondants who were exploratory and listened to music while working showed the largest correlations. 

Among the Mental Health Symptoms, Anxiety showed the best relationship with Music effects.

Interesting note about genres- Those that listened to R&B, Gospel, and Pop reported improved better mental health symptoms. I find it interesting that these genres are among the top becuase they're considered to be the most vocal heavy genres of music.

```python
corr_matrix['Total Mental Health Score'].sort_values(ascending = False)[5:]
```

```python
corr_matrix['Age'].sort_values(ascending = False)
```

Total Mental Health Score told a similar story:

Listening to a lot of music per day are correlated strong to to poor mental health scores

No surprise again that Age is negatively correlated to poor mental health scores

```python
df_genre_anxiety = df.groupby('Fav genre')['Anxiety'].mean().sort_values()
df_genre_depression = df.groupby('Fav genre')['Depression'].mean().sort_values()
df_genre_insomnia = df.groupby('Fav genre')['Insomnia'].mean().sort_values()
df_genre_ocd = df.groupby('Fav genre')['OCD'].mean().sort_values()
df_genre_total_score = df.groupby('Fav genre')['Total Mental Health Score'].mean().sort_values()
df_genre_music_effects = df.groupby('Fav genre')['Music effects'].mean().sort_values()

scoring_df = pd.concat([df_genre_anxiety, df_genre_depression, df_genre_insomnia, df_genre_ocd, df_genre_total_score, df_genre_music_effects], axis = 1).sort_values(by = 'Total Mental Health Score')
scoring_df
```

ADDITIONAL CONCLUSIONS:

It's important to note that music effects and mental health score are tell two different stories. mental health score is a rating by the respondant of their well being and Music effects is the perception by that respondant that Music improved their Mental Health. 

Below is a chart that breaks down the relationship between Mental Health Score and Music effects. It's interesting that people who reported that music has no effect had the best mental health scores. 

I found it interesting that Older people who generally listen to rock and folk the most, are also the least prone to depression even though rock and folk showed a high correlation to poor mental health score. I find this to be a pretty profound statistic.

It's safe to say that respondants who were exploratory in their music listening, generally had improvements in their mental health. There are a few studies out there that would say listening to new music improved brain neuralplasicity , but I'm not convinced this data can prove that music improves mental health. There's a relationship between Hours of music listening per day and poor mental health score.

It's worth noting that people's behavior could be explained by the fact that they believe they're mental health is poor. If I re-ran this survey, I would ask "Does your mental health play a factor in your music choices?". For example, Video game and lofi music (known to be soft listening music) are generally correlating to mental health score. I'm curious to know if people are listening to Lofi and Video Game music because their Mental Health is poor or if listening to video game and lofi music is causing their mental health score to be poor.
